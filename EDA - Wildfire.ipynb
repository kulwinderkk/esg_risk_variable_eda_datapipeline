We have dimensions (rlon = 424 , rlat = 412 , time = 365 ). Each file has dimension of rlon (longitude in rotated pole grid) and rlat (latitude in rotated pole grid) just like x and y coordinates and they make up the latitude and longitudes.

There is only one variable fwi-daily-proj i.e Fire Weather Index, and the data has daily frequency and time period covered is form 2021 to 2049.

```python
# Extracting longitude, latitude, and FWI values
lon = ds1['lon']
lat = ds1['lat']
fwi = ds1['fwi-daily-proj']


# Plotting the FWI values
plt.figure(figsize=(10, 6))
plt.contourf(lon, lat, fwi[0], cmap='hot', levels=20)  # Assuming you want to plot the first time step
plt.colorbar(label='FWI')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'Fire Weather Index on {ds1['time'][0].values}')
plt.show()
```

## Merging all the NC files in Folder and analysisng it

```python
ds_merged=xr.open_mfdataset(os.path.join(directory_path, '*.nc'), parallel=True)
```

```python
ds_merged
```

```python
ds_merged['lat'].shape
```

```python
ds_merged.info()
```

```python
ds_merged['time'].values
```

```python
# Resample the data to yearly frequency and calculate the mean
fwi_yearly_mean = ds_merged['fwi-daily-proj'].resample(time='1YE').mean(dim='time')
```

```python
fwi_yearly_values=fwi_yearly_mean.values
```

```python
fwi_yearly_values.shape
```

```python
plt.figure(figsize=(20,8))
plt.bar(fwi_yearly_mean['time'].dt.year, fwi_yearly_values[:, 0, 0])
plt.xticks(fwi_yearly_mean['time'].dt.year)
plt.title("Yearly mean Fire Index")
plt.xlabel("Year")
plt.ylabel("Fire Index")
plt.show()
```

```python
fwi_yearly_mean['time']
```

```python
# Extracting longitude, latitude, and FWI values
lon = ds_merged['lon']
lat = ds_merged['lat']
fwi = ds_merged['fwi-daily-proj']


# Plotting the FWI values
plt.figure(figsize=(10, 6))
plt.contourf(lon, lat, fwi[400], cmap='hot', levels=20)  # Assuming you want to plot the first time step
plt.colorbar(label='FWI')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'Fire Weather Index on {ds_merged['time'][780].values}')
plt.show()
```

```python
ds_merged['fwi-daily-proj']
```

```python
# Plotting FWI over time at specific locations
plt.figure(figsize=(10, 6))
ds_merged['fwi-daily-proj'][0:365].isel(rlon=0, rlat=0).plot(label='FWI at rlon=0, rlat=0')
plt.xlabel('Time')
plt.ylabel('FWI')
plt.title('Daily Fire Weather Index at Specific Loacation')
plt.legend()
plt.show()
```

```python
# Specify the file path
file_path = 'wildfire_index_europe.nc'

# Check if the file exists
if os.path.exists(file_path):
    # If the file exists, delete it
    os.remove(file_path)
ds_merged.to_netcdf('wildfire_index_europe.nc')
```
